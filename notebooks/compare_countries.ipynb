```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for statistical analysis
from scipy.stats import f_oneway

```

```python
benin = pd.read_csv('../data/benin_clean.csv')
sierra_leone = pd.read_csv('../data/sierra_leone_clean.csv')
togo = pd.read_csv('../data/togo_clean.csv')

# Check the first few rows of each to confirm they loaded correctly
print("Benin data:")
display(benin.head())

print("Sierra Leone data:")
display(sierra_leone.head())

print("Togo data:")
display(togo.head())
```

```python
plt.figure(figsize=(10,6))

# Combine data into one DataFrame with a new column 'Country'
benin['Country'] = 'Benin'
sierra_leone['Country'] = 'Sierra Leone'
togo['Country'] = 'Togo'

combined = pd.concat([benin, sierra_leone, togo], ignore_index=True)

sns.boxplot(x='Country', y='GHI', data=combined)
plt.title('Comparison of GHI Across Countries')
plt.ylabel('Global Horizontal Irradiance (GHI)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: loading your cleaned CSV files again
benin_df = pd.read_csv('../data/benin_clean.csv')
sierra_leone_df = pd.read_csv('../data/sierra_leone_clean.csv')
togo_df = pd.read_csv('../data/togo_clean.csv')

# Add 'Country' column to each DataFrame
benin_df['Country'] = 'Benin'
sierra_leone_df['Country'] = 'Sierra Leone'
togo_df['Country'] = 'Togo'

# Combine into a single DataFrame
df = pd.concat([benin_df, sierra_leone_df, togo_df], ignore_index=True)

# Now plot boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='GHI')
plt.title('GHI Distribution by Country')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='DNI')
plt.title('DNI Distribution by Country')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='DHI')
plt.title('DHI Distribution by Country')
plt.show()
```

```python
variables = ['GHI', 'DNI', 'DHI']
for var in variables:
    plt.figure(figsize=(12, 6))
    sns.histplot(data=df, x=var, hue='Country', kde=True, bins=50)
    plt.title(f'{var} Distribution by Country')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.legend(title='Country')
    plt.show()
```

```python
# Group by Country and calculate mean, median, min, max, and std
summary_stats = df.groupby('Country')[['GHI', 'DNI', 'DHI']].agg(['mean', 'median', 'min', 'max', 'std'])
summary_stats
```
